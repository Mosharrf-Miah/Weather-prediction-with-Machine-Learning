```python
# removing date from the pleasant dataset

pleasant = pleasant.drop(['DATE'], axis=1)
```

```python
#Create an 'X' matrix
X = df2
y = pleasant
```

```python
print(X.shape)
print(y.shape)
```

```python
### Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Run Decision Tree classifier
clf = DecisionTreeClassifier(criterion='gini', min_samples_split=2)
clf.fit(X_train, y_train)
figure(figsize=(10, 20))
tree.plot_tree(clf, feature_names=X_train.columns)  # Use training data columns
```

```python
#cross validation method
y_pred_train = clf.predict(X_train)
print('Train accuracy score: ',cross_val_score(clf, X_train, y_train, cv = 3, scoring='accuracy').mean())
```

```python
#cross validation method
y_pred_train = clf.predict(X_train)
print('Test accuracy score: ',cross_val_score(clf, X_test, y_test, cv = 3, scoring='accuracy').mean())
```

```python
locations = list({x.split("_")[0] for x in df2.columns if x not in ["MONTH", "DATE"]})
locations.sort()
```

```python
fig, ax = plt.subplots(3, 5, figsize=(25, 15))
count = 0

for i in range(3):
    for j in range(5):
        name = locations[count]
        target_column = name + "_pleasant_weather"
        
        y = pleasant[target_column]
        X_train, X_test, y_train, y_test = train_test_split(df2, y, random_state=42)
        
        clf = DecisionTreeClassifier()
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)
        
        cm = confusion_matrix(y_test, y_pred)
        sns.heatmap(cm, annot=True, fmt='g', ax=ax[i, j])
        ax[i, j].set_xlabel('Predicted')
        ax[i, j].set_ylabel('True')
        ax[i, j].set_title(name + ' Matrix')
        
        count += 1
```
